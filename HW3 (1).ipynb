```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import linalg as LA

Markov_Transition_matrix=np.zeros((101,101))
for i in range(1,101):
    Markov_Transition_matrix[i:i+6,i-1]=1/6


Markov_Transition_matrix[100,95:100]+=np.linspace(1/6,5/6,5)
Markov_Transition_matrix[100,100]=1
```

```python
print(Markov_Transition_matrix[:,100])
```

```python
count=0
#print(Transition_matrix.shape[1])
for i in range(Markov_Transition_matrix.shape[1]):
    column=Markov_Transition_matrix[:,i]
    #print(column)
    column_sum=0
    for x in column:
        column_sum+=x
    #column_sum=np.sum(column)
    #print("Column sum="+str(column_sum))
    if(not(column_sum>=0.9999999999999999)):
        print('The martix is not markov matrix')
        break
    else:
        count+=1
    #print("Number of Sum1 columns="+str(count))
if(count==Markov_Transition_matrix.shape[1]):
    count=0
    for i in Markov_Transition_matrix:
        for j in i:
            if (j>=0 and j<=1):
                count+=1
    
    if count==Markov_Transition_matrix.shape[0]*Markov_Transition_matrix.shape[1]:
        eig_values=LA.eigvals(Markov_Transition_matrix)
        if max(eig_values.tolist())==1:
            print('The matrix is a markov matrix')
```

```python
x0=np.zeros(101)
x0[0]=1
print('Initial State- ')
print(x0)

x1=Markov_Transition_matrix.dot(x0)
print('First move')
print(x1)
```

```python
def compute(k,x0):
    x_k=LA.matrix_power(Markov_Transition_matrix,k).dot(x0)
    return x_k

k=int(input('Enter the number of moves'))
prob_vector=compute(k,x0)
print(prob_vector)
```

<b>Explanation</b><br>
The steady state vector is the goal state, i.e, when the player has reached the goal position. Since, we know that player will definitely reach the goal posistion in 100 moves, I am computing the steady state vector using the function above for k=100.   

```python
x_100=compute(100,x0)
print('Steady state vector-')
print(x_100)
```

<b>Eigen values and vectors using numpy Linalg</b>

```python
eig_va,eig_v=LA.eig(Markov_Transition_matrix)
print('Eigen_value')
print(eig_va[0])
print('Eigen vector- ')
print(eig_v[:,0])
```

The first eigen vector and steady state vector computed are almost equal. You can see that the values of each entry differ by an order of 10^-5.

```python
prob_10=compute(10,x0)[100]
prob_100=compute(100,x0)[100]

print('The probability that the game finishes in 10 moves- '+str(prob_10))
print('The probability that the game finishes in 100 moves- '+str(prob_100))
print('The probability that the game finishes after 10 moves- '+str(1-prob_10))
print('The probability that the game finishes after 100 moves-'+str(1-prob_100))
```

<b>Explanation</b>

In this version of the game, the minimum number of moves that a palyer can have is 1. So, eventually the game will be over, that is the player will reach tile square 100. In ten moves the maximum number that the player can reach is 60, and in hundred moves the minimum number that the player can reach is 100. This explains the probabilities observed above.

```python
k=int(input('Enter the number of moves'))
cumulative_probabilities=[]
def plot_prob(k):
    for i in range(1,k+1):
        cumulative_probabilities.append(compute(i,x0)[100])
    moves=np.linspace(1,k,k)
    plt.plot(moves,cumulative_probabilities)
    plt.ylabel('probability of finishing in less than k moves')
    plt.xlabel('moves(k)')
plot_prob(k)
```

```python
def exact_prob(k):
    curr_prob=compute(k,x0)[100]
    prev_prob=compute(k-1,x0)[100]
    exact_prob=curr_prob-prev_prob
    return exact_prob

print('Probability of finishing the game in exactly 50 moves is- '+str(exact_prob(50)))

    
```

```python
k=int(input('Enter the number of moves'))
exact_probabilities=[]
def plot_prob(k):
    for i in range(1,k+1):
        exact_probabilities.append(exact_prob(i))
    moves=np.linspace(1,k,k)
    plt.plot(moves,exact_probabilities)
    plt.xlabel('moves')
    plt.ylabel('Probability of finishing in k moves')
plot_prob(k)
```

```python
expected_value=0
for i in range(1,len(exact_probabilities)+1):
    expected_value+=i*exact_probabilities[i-1]
print('The expected value is- '+str(expected_value))
```

```python
ladders = [(1,39), (4,14), (9,31), (28,84), (36,44),
           (51,67), (80,100), (71,91)]
snakes = [(16,6), (47,26), (49,11), (56,53), (64,60),
          (92,73), (95,75), (98,78)]
trans = ladders + snakes
Transition_matrix = np.zeros((101, 101))
for i in trans:
    Transition_matrix[i[1],i[0]]=1

for j in range(0,101):
    column=Transition_matrix[:,j]
    count=0
    for k in column:
        if k==0:
            count+=1
    if count==len(column):
        Transition_matrix[j,j]=1
```

```python
x0=np.zeros(101)
x0[0]=1
print('Initial State- ')
print(x0)


M=Markov_Transition_matrix.dot(x0)
x1=Transition_matrix.dot(M)
print('After first move')
print(x1)
```

<b>Explanation</b><br>
The steady state vector is the goal state, i.e, when the player has reached the goal position.Since, we know that the player would may reach the goal state depending on the roll of the dice and position of chutes and ladders, I am computing the steady state vector using the above function for k=400. 

```python
def compute_2(k,x0):
    x_k=LA.matrix_power(Transition_matrix.dot(Markov_Transition_matrix),k).dot(x0)
    return x_k

k=int(input('Enter the number of moves'))
prob_vector=compute_2(k,x0)
print(prob_vector)
```

```python
x_100=compute_2(400,x0)
print('Steady state vector-')
print(x_100)
```

```python
prob_10=compute_2(10,x0)[100]
prob_100=compute_2(100,x0)[100]

print('The probability that the game finishes in 10 moves- '+str(prob_10))
print('The probability that the game finishes in 100 moves- '+str(prob_100))
print('The probability that the game finishes after 10 moves- '+str(1-prob_10))
print('The probability that the game finishes after 100 moves-'+str(1-prob_100))
```

<b>Explanation</b>

In this version of the game, the minimum number of squares which a player can cover in each move is 1 and the maximum is 6. But, there are constraints in the form of chutes and ladders- these two constraints can either increase or decrease the number of steps required to reach the goal state, depending on the output of the dice. In the previous, we established that a player cannot reach the goal state in 10 moves as the maximum steps he/she can cover is 60. Also, we established that the minimum number of steps a player can cover in 100 steps is 100, which is equivalent to the length of the board. But, here due to the constraints, it is possible to reach the goal state in 10 moves, also it's possible to not reach the goal state in 100 moves. I hope this explains the probabilities observed above.
