```python
import numpy as np
import scipy.io as sio
```

### Load data from .mat file

```python
mat_contents = sio.loadmat('ash85_data.mat') # ash219_data, ash85_data
data1 = mat_contents['x']
#print(data1)
#load the other matrix ash219_data'
#new_mat_contents=sio.loadmat('ash219_data.mat')
#print(new_mat_contents.items())
#data2=new_mat_contents['x']
```

### Return a dense ndarray representation of this matrix.

```python
A = data1.toarray() # or todense()
print(A.shape)
print(A)
```

```python
from scipy.linalg import lu_factor, lu_solve
# Task 1: LU factorization
# define a right hand vector b with all ones
# use lu_factor and lu_solve to solve Ax=b for both inputs
# solve for both ash219_data, ash85_data
def create_matrix_b(rows_of_b):
    b=[]
    for i in range(rows_of_b):
        b.append(1)
    return b

output_vector=create_matrix_b(A.shape[1]) #Create an output vector based on the shape of the input vector
factored_matrix_A=lu_factor(A) #Factor the matrix into LU
print('The factored matrix is- ')
print(factored_matrix_A)
solution=lu_solve(factored_matrix_A,output_vector,trans=0)
print('The solution is-')
print(solution)
```

```python
# Task 2: verify the correctness of your solutions. 
# In the verification, you can use 
# numpy.allclose function that allows some tolerance
# verify the correctness of solutions for both ash219_data, ash85_data
#x=np.linalg.solve(A,output_vector) 
#print(x)
#if np.allclose(solution,x,atol=1.0):
#    print('Your solution is correct')
#else:
#    print('Your solution is wrong')
manual_output_vector=np.matmul(A,solution) #Generates the output_vector based on the solution obtained above. 
#Check if the components of the output vector generated above are within the tolerance limit
if np.allclose(manual_output_vector,output_vector,atol=1.0):
    print('Your solution is correct and all the elements are within the tolerance limit')
else:
    print('Your solution is wrong')
    
```
